```python
import pandas as pd

# Load the CSV, ensuring commas within quotes are handled
df2 = pd.read_csv('dog.csv', quotechar='"', encoding='utf-8')

# Check the first few rows
print(df2.head())
print(df2.columns)
```

```python
df2.head(10)
```

```python
column_name = 'curated_breed_name'

# Count non-NaN values
non_nan_count = df2[column_name].notna().sum()

# Filter rows where the column is not NaN
non_nan_rows = df2[df2[column_name].notna()]

# Print results
print(f"Number of non-NaN values in '{column_name}': {non_nan_count}")
print(f"\nRows with non-NaN values in '{column_name}':")
print(non_nan_rows)
```

```python
# Split the 'curated_breed_name' column on commas and explode into separate rows
df2['curated_breed_name'] = df2['curated_breed_name'].str.split(',')
df_expanded = df2.explode('curated_breed_name')

# Remove any leading/trailing whitespace from breed names
df_expanded['curated_breed_name'] = df_expanded['curated_breed_name'].str.strip()

# Check the result
print("Expanded DataFrame:")
print(df_expanded.head(10))
print("Shape:", df_expanded.shape)
```

```python
# Columns to remove (adjust based on your needs)
columns_to_remove = ['biomarkerUsage', 'sample', 'cellLine', 'roc', 'km', 'cox', 'gene', 'traitAssociation', 'pmid' ]

# Remove the columns
df_expanded = df_expanded.drop(columns=columns_to_remove)

# Display the result
print("Columns after removal:", df_expanded.columns.tolist())
print("First few rows:")
print(df_expanded.head())

# Optional: Save the updated DataFrame
df_expanded.to_csv('dogs_cleaned.csv', index=False)
```

```python
breed_names = df['breed_name'].unique()
curated_breed_names = df_expanded['curated_breed_name'].dropna().unique()

# Find matching values
matching_breeds = set(breed_names).intersection(curated_breed_names)

# Print results
print(f"Number of unique breed_name values: {len(breed_names)}")
print(f"Number of unique curated_breed_name values: {len(curated_breed_names)}")
print(f"Number of matching breeds: {len(matching_breeds)}")
print("Matching breeds:", sorted(matching_breeds))
```

```python

# Find matching and non-matching values
matching_breeds = set(breed_names).intersection(curated_breed_names)
only_in_breed_name = set(breed_names) - set(curated_breed_names)
only_in_curated = set(curated_breed_names) - set(breed_names)

# Print results
print(f"Unique breed_name values: {len(breed_names)}")
print(f"Unique curated_breed_name values (excluding NaN): {len(curated_breed_names)}")
print(f"Matching breeds ({len(matching_breeds)}): {sorted(matching_breeds)}")
print(f"\nBreeds only in 'breed_name' ({len(only_in_breed_name)}): {sorted(only_in_breed_name)}")
print(f"Breeds only in 'curated_breed_name' ({len(only_in_curated)}): {sorted(only_in_curated)}")

# Additional info
print(f"\nNaN in 'breed_name': {df['breed_name'].isna().sum()}")
print(f"NaN in 'curated_breed_name': {df2['curated_breed_name'].isna().sum()}")
```

```python
combined_df_left = pd.merge(df, df_expanded, 
                            left_on='breed_name', 
                            right_on='curated_breed_name', 
                            how='left')
print("\nLeft Merge (All from df):")
print("Shape:", combined_df_left.shape)
print(combined_df_left.head())
```

```python
print("Merged columns:", combined_df_left.columns.tolist())
```
